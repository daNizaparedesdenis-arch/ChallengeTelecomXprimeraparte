#ðŸ”§ TransformaciÃ³n

```python
telecomX_customer_df = pd.json_normalize(datos_telecomX["customer"])
telecomX_phone_df = pd.json_normalize(datos_telecomX["phone"])
telecomX_internet_df = pd.json_normalize(datos_telecomX["internet"])
telecomX_account_df = pd.json_normalize(datos_telecomX["account"])

datos_telecomX_final = pd.concat([datos_telecomX.drop(columns=['customer','phone', 'internet', 'account']),
                                 telecomX_customer_df,
                                 telecomX_phone_df,
                                 telecomX_internet_df,
                                 telecomX_account_df],
                                axis=1)
```

```python
datos_telecomX_depurada = datos_telecomX_final.drop(columns=['PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 'PaperlessBilling'])
datos_telecomX_depurada.head(5)
```
